We aim to identify the most suitable locations for a new hospital in San Diego by
analyzing spatial data on school lands, income levels by county, and road traffic injuries.

healthcare services, aiming to propose locations that would optimize accessibility,
community need, and safety. The expected audience includes healthcare planners, local
government, and community health organizations, who would benefit from insights into
strategic hospital placement to enhance healthcare access and emergency response
efficiency

```python
census = gpd.read_file("Census/CENSUS_TRACTS_2020.shp")
merged_df = pd.merge(income, census, right_on='TRACT', left_on='tract', how='inner')

income_gdf = gpd.GeoDataFrame(merged_df)
income_gdf = income_gdf[income_gdf["YEAR"] == 2016].reset_index()
income_gdf = income_gdf.to_crs("EPSG:4326")


income_group_values = {
    "Less than $15,000": 7500,
    "15,000 to 29,999": (15000 + 29999) / 2,
    "30,000 to 44,999": (30000 + 44999) / 2,
    "45,000 to 59,999": (45000 + 59999) / 2,
    "60,000 to 74,999": (60000 + 74999) / 2,
    "75,000 to 99,999": (75000 + 99999) / 2,
    "100,000 to 124,999": (100000 + 124999) / 2,
    "125,000 to 149,999": (125000 + 149999) / 2,
    "150,000 to 199,999": (150000 + 199999) / 2,
    "$200,000 or more": 200000
}


income_gdf['INCOME_MIDPOINT'] = income_gdf['INCOME GROUP'].map(income_group_values)

income_gdf['TOTAL_INCOME'] = income_gdf['HOUSEHOLDS'] * income_gdf['INCOME_MIDPOINT']


tract_income = income_gdf.groupby('tract').agg(
    TOTAL_HOUSEHOLDS=('HOUSEHOLDS', 'sum'),
    TOTAL_INCOME=('TOTAL_INCOME', 'sum')
).reset_index()


tract_income['AVERAGE_INCOME'] = tract_income['TOTAL_INCOME'] / tract_income['TOTAL_HOUSEHOLDS']


low_high_threshold = tract_income['AVERAGE_INCOME'].quantile(1/3)
high_average_threshold = tract_income['AVERAGE_INCOME'].quantile(2/3)


tract_income['INCOME_CATEGORY'] = pd.cut(tract_income['AVERAGE_INCOME'],
                                         bins=[tract_income['AVERAGE_INCOME'].min()-1, 
                                         low_high_threshold, high_average_threshold, 
                                         tract_income['AVERAGE_INCOME'].max()],
                                         labels=['Low', 'Average', 'High'])


unique_geometries = income_gdf.dissolve(by='tract', aggfunc='first')


tract_income_geo = pd.merge(tract_income, unique_geometries[['geometry']], left_on='tract', right_index=True, how='left')


tract_income_gdf = gpd.GeoDataFrame(tract_income_geo, geometry='geometry', crs="EPSG:4326")


tract_income_gdf.head()
```

```python
base = sd_county_boundary.to_crs("EPSG:4326").plot(figsize=(10, 10), color='none', edgecolor='black')


tract_income_gdf.plot(ax=base, column='INCOME_CATEGORY', legend=True, 
                      edgecolor='black', cmap='RdYlGn')

plt.show()
```

### However, to solidify our findings, we employed a more rigorous statistical method by conducting a chi-square test. This test is designed to analyze the strength of association between two categorical variables. In our case, these were the income levels of different areas and the frequency of traffic crashes within those areas. The results of this test, indicated by a chi-square value close to 0 and a P-value near 1, suggested that there is no strong statistical correlation between the income levels and the incidence of traffic accidents. This outcome led us to the decision to exclude income level data from our criteria for hospital placement, focusing instead on more direct indicators of healthcare needs.

## 2) Hot Spots and Cold Spots (Spatial autocorrelation)

```python
y = Sandiego_Crash['VICTIM_DEGREE_OF_INJURY']
yb = y <= y.median()
labels = ["1 Low", "0 High"]
yb = [labels[i] for i in 1*yb] 
Sandiego_Crash['yb'] = yb
```

```python
total_bounds = sd_county_boundary.total_bounds
minx, miny, maxx, maxy = total_bounds

# Create a grid within the bounding box
x_range = np.linspace(minx, maxx, 50) # 50 points for granularity
y_range = np.linspace(miny, maxy, 50)
grid_cells = []

for x in range(len(x_range)-1):
    for y in range(len(y_range)-1):
        grid_cells.append(Polygon([(x_range[x], y_range[y]), 
                                   (x_range[x+1], y_range[y]), 
                                   (x_range[x+1], y_range[y+1]), 
                                   (x_range[x], y_range[y+1])]))

# Step 2: Map crash points to the grid
Sandiego_Crash['geometry'] = Sandiego_Crash.apply(lambda row: Point(row['POINT_X'], row['POINT_Y']), axis=1)
Sandiego_Crash_gdf = gpd.GeoDataFrame(Sandiego_Crash, geometry='geometry')

# Color crashes based on severity for visualization
Sandiego_Crash_gdf['color'] = Sandiego_Crash_gdf['yb'].map({"0 Low": "blue", "1 High": "red"})

# Step 3: Plotting
fig, ax = plt.subplots(figsize=(12, 12))
sd_county_boundary.plot(ax=ax, color='none', edgecolor='black')

# Plot grid cells
for cell in grid_cells:
    xs, ys = cell.exterior.xy
    ax.fill(xs, ys, alpha=0.5, fc='none', ec='gray')

# Correcting the labels based on severity
severity_labels = {"1 Low": "Less Severe Injury", "0 High": "More Severe Injury"}
Sandiego_Crash_gdf['severity_label'] = Sandiego_Crash_gdf['yb'].map(severity_labels)

# Plot crash points with corrected severity labels and ensure color coding is correct
for label, df_group in Sandiego_Crash_gdf.groupby('severity_label'):
    color = 'red' if label == "More Severe Injury" else 'blue'
    df_group.plot(ax=ax, color=color, markersize=5, label=label)

# Adjustments for schools plotting if included, same as before, no changes needed

plt.xlim(minx, maxx)
plt.ylim(miny, maxy)
plt.title("San Diego Traffic Crash Analysis")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.show()
```

### Shifting our attention to identifying hotspots more accurately, we opted for a novel approach by overlaying an artificial grid on the map of San Diego County. This method allowed us to systematically dissect the geographic area into smaller, uniformly sized cells, facilitating a more granular analysis of traffic accident occurrences. Through this technique, we were able to identify hotspots—grid cells exhibiting a statistically significant higher number of traffic accidents than the average. In contrast, cells with significantly fewer accidents were designated as coldspots. Furthermore, we assigned additional priority to cells encompassing multiple schools, recognizing the heightened need for emergency healthcare services in areas with a dense population of young residents.

```python
grid_cells_gdf = gpd.GeoDataFrame(geometry=grid_cells)
joined_gdf = sjoin(Sandiego_Crash_gdf, grid_cells_gdf, how="inner", op='intersects')
crash_counts = joined_gdf.groupby('index_right').size()

grid_cells_gdf['crash_counts'] = grid_cells_gdf.index.map(crash_counts).fillna(0).astype(int)
grid_cells_gdf['crash_counts'] = grid_cells_gdf['crash_counts'].fillna(0).astype(float)

w = lps.weights.Queen.from_dataframe(grid_cells_gdf)
w.transform = 'r'

g_local = G_Local(grid_cells_gdf['crash_counts'], w, star=True, n_jobs=1)
grid_cells_gdf['Gi*_ZScore'] = g_local.Zs

fig, ax = plt.subplots(figsize=(12, 12))
grid_cells_gdf.plot(column='Gi*_ZScore', cmap='coolwarm', legend=True, ax=ax)

Sandiego_Crash_gdf.plot(ax=ax, color='red', markersize=5, label='Crashes')
for idx, row in Sandiego_school.iterrows():
    school_centroid = row['geometry'].centroid
    for cell in grid_cells_gdf:
        ax.plot(school_centroid.x, school_centroid.y, marker='o', color='black', markersize=5, label='School' if idx == 0 else "")
        break

sd_county_boundary.plot(ax=ax, color='none', edgecolor='black', linewidth=2)

plt.legend()
plt.xlim(minx, maxx)
plt.ylim(miny, maxy)
plt.title('Refined Hotspot Analysis of Crashes with Schools in San Diego')
plt.show()
```

```python
median_injury_degree = Sandiego_Crash_gdf['VICTIM_DEGREE_OF_INJURY'].median()

# Filter crashes where injury degree is greater than the median
high_severity_crashes_gdf = Sandiego_Crash_gdf[Sandiego_Crash_gdf['VICTIM_DEGREE_OF_INJURY'] > median_injury_degree]

# Proceed with your spatial analysis using high_severity_crashes_gdf
# Example: Spatial join high severity crashes with grid cells
joined_gdf = sjoin(high_severity_crashes_gdf, grid_cells_gdf, how="inner", op='intersects')
crash_counts = joined_gdf.groupby('index_right').size()

# Update grid_cells_gdf with crash counts, then proceed as before
grid_cells_gdf['crash_counts'] = grid_cells_gdf.index.map(crash_counts).fillna(0).astype(float)

# Create spatial weights, perform G_Local analysis
w = lps.weights.Queen.from_dataframe(grid_cells_gdf)
w.transform = 'r'
g_local = G_Local(grid_cells_gdf['crash_counts'], w, star=True, n_jobs=1)

# Update grid_cells_gdf with the results
grid_cells_gdf['Gi*_ZScore'] = g_local.Zs

# Visualization (omitting repeated code for brevity)
fig, ax = plt.subplots(figsize=(12, 12))
grid_cells_gdf.plot(column='Gi*_ZScore', cmap='coolwarm', legend=True, ax=ax)
sd_county_boundary.plot(ax=ax, color='none', edgecolor='black', linewidth=2)
#Sandiego_Crash_gdf.plot(ax=ax, color='red', markersize=5, label='Crashes')
plt.legend()
plt.xlim(minx, maxx)
plt.ylim(miny, maxy)
plt.title('Refined Hotspot Analysis of Crashes with Schools in San Diego')
plt.show()
```

```python
def decide_hospital_placement(z_score):
    if z_score > 8:
        return 1.0
    elif z_score > 6:
        return 0.35
    elif z_score > 4:
        return 0.25
    elif z_score > 2:
        return 0.2
    elif z_score > 0:
        return 0.1
    else:
        return 0.005  # No hospital for lower Z-Scores
np.random.seed(48)
# Apply the decision rule to each grid cell and determine if a hospital will be placed
grid_cells_gdf['hospital_chance'] = grid_cells_gdf['Gi*_ZScore'].apply(decide_hospital_placement)
grid_cells_gdf['place_hospital'] = np.random.binomial(1, grid_cells_gdf['hospital_chance'])

# Calculate centroid for potential hospital locations
grid_cells_gdf['hospital_location'] = grid_cells_gdf[grid_cells_gdf['place_hospital'] == 1].geometry.centroid

# Visualization
def plot_estimated_hospitals_within_boundary(ax, hospital_locations, county_boundary):
    for location in hospital_locations.dropna():
        # Check if the hospital location is within the county boundary and outside the restricted longitude range
        if county_boundary.contains(location).any() and not (-116.9 <= location.x):
            ax.plot(location.x, location.y, marker='*', color='Yellow', markersize=10, label="Hospital (est.)")

# Visualization
fig, ax = plt.subplots(figsize=(12, 12))

# Plot hotspots and coldspots
grid_cells_gdf.plot(column='Gi*_ZScore', cmap='coolwarm', legend=True, ax=ax)

# Plot estimated hospital locations within the county boundary
plot_estimated_hospitals_within_boundary(ax, grid_cells_gdf['hospital_location'], sd_county_boundary)

# Plot San Diego county boundary for reference
sd_county_boundary.plot(ax=ax, color='none', edgecolor='black', linewidth=2)

# Adjust legend to show the hospital marker only once
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))  # Removes duplicate labels/handles
plt.legend(by_label.values(), by_label.keys())

plt.xlim(minx, maxx)
plt.ylim(miny, maxy)
plt.title('Hospital Placement Based on Traffic Crash Hotspots within San Diego County')
plt.show()
```

### Utilizing Z-scores derived from Hotspots analysis to guide the placement of hospitals in San Diego County based on traffic crash data represents a sophisticated approach to urban planning and public health strategy. This method involves a statistical analysis where Z-scores are calculated for each area based on the frequency of traffic accidents, essentially measuring how many standard deviations an element is from the mean. A higher Z-score indicates a greater deviation, which, in this context, signifies a hotspot for traffic crashes. By targeting these areas, the strategy prioritizes the establishment of healthcare facilities where they are most needed, ensuring that emergency medical services are readily accessible in locations with a higher likelihood of accidents. This proactive approach not only optimizes resource allocation but also has the potential to significantly reduce response times in critical situations, thereby improving patient outcomes.
